```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4]
y = [5,6,7,8]

plt.plot(x,y) # Line plot
plt.show() # It is not necessary in jupyter lab but when we work on any other editor it is necessary for showing the plot.
```

```python
# Movies Data

# Oscar Winner Movies
oscar_movies = [
    "The Dark Knight", 
    "The Hurt Locker",  
    "The King's Speech", 
    "The Artist",     
    "Argo"
]
oscar_years = [2008, 2009, 2010, 2011, 2012]
oscar_revenue = [1005, 170, 427, 133, 232]  # in $M

plt.plot(oscar_years, oscar_revenue)
plt.title("Oscar Years vs Revenue(in $M)")
plt.xlabel("Years")
plt.ylabel("Revenue(in $M)")
```

```python
non_oscar_movies = ["Slumdog Millionaire", "Avatar", "Inception", "Hugo", "Lincon"]
non_oscar_revenue = [3456, 8765, 567, 3789, 4565]

plt.plot(oscar_years, oscar_revenue, label = "Oscar Movie")
plt.plot(oscar_years, non_oscar_revenue, label = "Non Oscar Movie")
plt.title("Oscar Years vs Oscar & Non-oscar revenue(in $M)")
plt.xlabel("Years")
plt.ylabel("Revenue(in $M)")

plt.legend(loc = "upper right") # This is for showing label written in plot() / location is not mandatory, by default it's position in best place.
```

```python
# Change marker, line and color of line in plot (format String)
plt.plot(oscar_years, oscar_revenue, "o--r", linewidth=5, label = "Oscar Movie") #Format String(marker, line, Color)
plt.plot(oscar_years, non_oscar_revenue, color = "green", linestyle="-.", label = "Non Oscar Movie") # Color change only, can use #code also
plt.title("Oscar Years vs Oscar & Non-oscar revenue(in $M)")
plt.xlabel("Years")
plt.ylabel("Revenue(in $M)")

plt.legend(loc = "upper right")
```

```python
plt.plot(oscar_years, oscar_revenue, label = "Oscar Movie")
plt.plot(oscar_years, non_oscar_revenue, label = "Non Oscar Movie")
plt.title("Oscar Years vs Oscar & Non-oscar revenue(in $M)")
plt.xlabel("Years")
plt.ylabel("Revenue(in $M)")

plt.legend(loc = "upper right")

plt.style.available # To show all available style (background)
plt.style.use("dark_background")
plt.grid() # Tu create greed in the background.
```

```python
# Tight_layout :

# Oscar Winner Movies
oscar_movies = [
    "The Dark Knight", 
    "The Hurt Locker",  
    "The King's Speech", 
    "The Artist",     
    "Argo"
]
oscar_years = [2008, 2009, 2010, 2011, 2012]
oscar_revenue = [1005, 170, 427, 133, 232]  # in $M

plt.style.use("default")
plt.plot(oscar_movies, oscar_revenue)
plt.title("Oscar Years vs Revenue(in $M)")
plt.xlabel("Moies name")
plt.ylabel("Revenue(in $M)")
```

```python
# In above the movies names are overlapped, To prevent this tight_layout is use.

# Oscar Winner Movies
oscar_movies = [
    "The Dark Knight",
    "The Hurt Locker",
    "The King's Speech",
    "The Artist",
    "Argo"
]

oscar_years = [2008, 2009, 2010, 2011, 2012]
oscar_revenue = [1005, 170, 427, 133, 232]  # in $M

plt.style.use("default")
plt.plot(oscar_movies, oscar_revenue)
plt.title("Oscar Years vs Revenue (in $M)")
plt.xlabel("Movie name")
plt.ylabel("Revenue (in $M)")

# Fix the spacing between labels and the plot
plt.tight_layout()

# Show the plot
plt.show()

# to save the plot in png format
plt.savefig("lineplot.png")
```
