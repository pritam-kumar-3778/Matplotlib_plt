```python
import matplotlib.pyplot as plt

expenses = ["Salaries", "Rent", "Marketing", "R&D", "Miscellaneous"]
amounts = [500, 150, 120, 100, 50]

explode_rnd =[0, 0, 0, 0.2, 0]

# Compare, comapny spent how much in each category
plt.pie(amounts, labels = expenses, autopct="%1.1f%%",
       wedgeprops = {
        "edgecolor":"red",
        "linewidth": 3,
        "linestyle": "--",
        # "fill": False  : Remove the color
        },
        explode = explode_rnd,
        startangle = 90, # in degree
        shadow = True
       )

plt.style.use("dark_background")
```

```python
autopct="%1.1f%%" : Show percentage of each slice
wedgeprops : define edgecolor, linestyle, linewidth, alpha, fill
explode : Heilight a slice by pulling it out
startangle : rotate the chart
shadow 
```
