```python
import matplotlib.pyplot as plt
```

```python
# Oscar Winner Movies
oscar_movies = [
    "The Dark Knight", 
    "The Hurt Locker",  
    "The King's Speech", 
    "The Artist",     
    "Argo"
]

non_oscar_movies = [
    "Slumdog Millionaire", 
    "Avatar", 
    "Inception", 
    "Hugo", 
    "Lincon"
]

oscar_revenue = [4005, 170, 427, 133, 232]  # in $M
non_oscar_revenue = [3456, 8765, 567, 3789, 4565]

year = [2008, 2009, 2010, 2011, 2012]

plt.bar(year, oscar_revenue)
plt.bar(year, non_oscar_revenue)

plt.title("Oscar VS Non-Oscar movie's revenue")
plt.xlabel("Years")
plt.ylabel("Revenue")
```

```python
# comparing data with bar charts
oscar_movies = ["The Dark Knight","The Hurt Locker","The King's Speech","The Artist","Argo"]
oscar_revenue = [1005, 170, 427, 133, 232]  # in $M

non_oscar_movies = ["Slumdog Millionaire","Avatar","Inception","Hugo","Lincoln"]
non_oscar_revenue = [378, 2788, 829, 185, 275]  # in $M

years = [2008, 2009, 2010, 2011, 2012]

# Plotting Bar Chart
import numpy as np
x = np.arange(len(years))
width = 0.4

plt.bar(x-width/2, oscar_revenue, width, label="Oscar")
plt.bar(x+width/2, non_oscar_revenue, width, label="Non-Oscar")

plt.xlabel("Years")
plt.ylabel("Revenue (in $M)")
plt.title("Yearly revenue of Oscar vs Non-Oscar movies")
plt.legend()
# plt.tight_layout()

plt.xticks(x, years) # xtiks() is for show x-axis labels
```
