```python
# axvline is use to draw a vertical line on a plot to show someting, like passing mark, average, ,passmarks or threesold.
```

```python
# axv line
marks = [
    25, 38, 42, 55, 61, 67, 72, 78, 83, 88,
    91, 95, 47, 53, 59, 64, 69, 74, 79, 85,
    92, 97, 33, 45, 58, 62, 71, 76, 81, 89
]

# custom bins
bins = [0, 30, 70, 90, 100]
plt.hist(
    marks,
    bins=bins,
    edgecolor="black",
    color="lightblue"
)

# line at passing marks - 33
plt.axvline(
    x=33,
    color="red",
    linestyle="--",
    linewidth=2,
    label="Pass Mark (33)"
)

plt.xlabel("Marks")
plt.ylabel("Number of Students")
plt.title("Distribution of Marks of 30 Students")
plt.grid(True)
```
