```python
import matplotlib.pyplot as plt
```

```python
days = ["Mon", "Tue", "Wed", "Thu", "Fri"]

cities = ["New york", "London", "Delhi", "Tokyo"]

temperetures = [
    [22,23,21,24,25], # New york
    [18,19,17,20,21], # London
    [30,32,31,33,34], # Delhi
    [25,26,24,27,28] # Tokyo
]

fig, axes = plt.subplots(2,2)

# city_no = 0
# for i in range(2) :  # rows (i) - 0,1
#     for j in range(2) :  # cols (j) - 0,1
#         axes[i][j].plot(days, temperetures[city_no], marker='o')
#         axes[i][j].set_title(cities[city_no])
#         axes[i][j].grid(True)
#         city_no += 1

ax = axes.flatten() # Flatten function makes 2D array to 1D array.
for i,ax in enumerate(ax): #enumerate function return index and value at index.
    ax.plot(days, temperetures[i], marker = 'o')
    ax.set_title("cities[i]")
    ax.grid(True)

fig.suptitle("Temperetures of the week over the week")

fig.supylabel("Tempereture")
fig.supxlabel("Days")

plt.tight_layout()
```
