```python
import matplotlib.pyplot as plt
```

```python
# Oscar Winner Movies
oscar_movies = [
    "The Dark Knight", 
    "The Hurt Locker",  
    "The King's Speech", 
    "The Artist",     
    "Argo"
]
oscar_years = [2008, 2009, 2010, 2011, 2012]
oscar_revenue = [1005, 170, 427, 133, 232]  # in $M

plt.bar(oscar_years, oscar_revenue)
plt.title("Years vs revenue of oscar movie")
plt.xlabel("year")
plt.ylabel("oscar_revenue in M$")
```

```python
# Now I want exact revenue value
oscar_movies = [
    "The Dark Knight", 
    "The Hurt Locker",  
    "The King's Speech", 
    "The Artist",     
    "Argo"
]
oscar_years = [2008, 2009, 2010, 2011, 2012]
oscar_revenue = [1005, 170, 427, 133, 232]  # in $M

plt.bar(oscar_years, oscar_revenue, color="purple")
plt.title("Years vs revenue of oscar movie")
plt.xlabel("year")
plt.ylabel("oscar_revenue in M$")

for i in range(len(oscar_years)):
    plt.text(oscar_years[i], oscar_revenue[i] + 20, str(oscar_revenue[i]), ha = "center") #ha for allignment
    # oscar_revenue[0] + 10 : + 10 (any value) is for increasing height of y axis for that the text is not touch the bar.

# By the increasing height of y axis the text touch the outer line so we have to extend the limitof y axis.
plt.ylim(0, max(oscar_revenue)+80) # 2 arguements : Lower limit and Upper limit)

# We can also extend limit of x axis by xlim(lower limit, Upper limit)
```
